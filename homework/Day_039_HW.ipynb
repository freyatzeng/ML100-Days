由於 L1 penalty規定的範圍'四四方方有菱有角' 所以最優解的係數會被剛好縮成0, 因此lasso可以實現對變量的選擇(因為係數=0的變量被篩掉了)

Ans: 因為LASSO在收斂時會將某些系數降為0,這些降為0的系數代表該feature可以被捨棄掉
